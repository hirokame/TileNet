```python
import numpy as np
import pandas as pd

import math

stateQ_list = []
n_stateQ_list = []

df = pd.read_csv('sampleNMQ_200-1.csv')
for i in range(200):
    stateQ = df[(df['s_row']==7) & (df['s_col']==3) & (df['episode']==i)]['stateQ'].max()
    n_stateQ = df[(df['s_row']==7) & (df['s_col']==3) & (df['episode']==i)]['n_stateQ'].max()
    if math.isnan(stateQ):
        stateQ = 0
    if math.isnan(n_stateQ):
        n_stateQ = 0
    stateQ_list.append(stateQ)
    n_stateQ_list.append(n_stateQ)
```

```python
import matplotlib.pyplot as plt
import numpy as np

t = np.arange(200)
y1 = np.array(stateQ_list)
y2 = np.array(n_stateQ_list)
sa = y2 - y1


fig, ax = plt.subplots(1,1,figsize=(15,8))
ax.fill_between(t,sa, color='#FDD79F')
ax.plot(t, y1, label='stateQ', color='#9698C8')
ax.plot(t, y2, label='n_stateQ', color='#F6878C')
# ax.set_title('learning_curve', fontsize=20)
ax.tick_params(labelsize=16)
ax.set_xlabel('Episode',fontsize=20)
ax.set_ylabel('Q value', fontsize=20)
ax.set_xlim(0,200)
ax.set_ylim(0,0.8)
plt.savefig('Q_nstate-Q_state.svg')

# ax.legend(fontsize='20')
```
