```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
file_name = pd.read_csv("sampleC-1.csv")
```

### STN, D1, D2のヒートマップ出力（epごと）

```python
# 初期のデータフレーム
stn_result = pd.DataFrame()
d1_result = pd.DataFrame()
d2_result = pd.DataFrame()

# i, jの範囲を設定
for i in range(2, 9):
    for j in range(2, 9):
        """STN"""
        stn_column_name = f'lc_hyper{i}{j}'  # lc_hyperij列名を作成
        stn_max_values = file_name.groupby(['episode'])[stn_column_name].max().reset_index()  # 最大値を計算し、インデックスをリセット
        stn_max_values.rename(columns={stn_column_name: f'max_{stn_column_name}'}, inplace=True)  # 列名を変更
        
        if stn_result.empty:
            stn_result = stn_max_values  # 初回はstn_resultに直接代入
        else:
            stn_result = pd.merge(stn_result, stn_max_values, on='episode')  # 横に結合
            
        """D1"""
        d1_column_name = f'D1{i}{j}'  # d1ij列名を作成
        d1_max_values = file_name.groupby(['episode'])[d1_column_name].max().reset_index()  # 最大値を計算し、インデックスをリセット
        d1_max_values.rename(columns={d1_column_name: f'max_{d1_column_name}'}, inplace=True)  # 列名を変更
        
        if d1_result.empty:
            d1_result = d1_max_values  # 初回はd1_resultに直接代入
        else:
            d1_result = pd.merge(d1_result, d1_max_values, on='episode')  # 横に結合
            
        """D2"""
        d2_column_name = f'D2{i}{j}'  # d2ij列名を作成
        d2_max_values = file_name.groupby(['episode'])[d2_column_name].max().reset_index()  # 最大値を計算し、インデックスをリセット
        d2_max_values.rename(columns={d2_column_name: f'max_{d2_column_name}'}, inplace=True)  # 列名を変更
        
        if d2_result.empty:
            d2_result = d2_max_values  # 初回はd2_resultに直接代入
        else:
            d2_result = pd.merge(d2_result, d2_max_values, on='episode')  # 横に結合
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
from matplotlib.colors import SymLogNorm
plt.rcParams["font.size"] = 10


epsilon = 1e-20
for ep in range(1, 10, 1):
    """STN"""
    print(f'STN ep={ep}')
    # 特定のepisode番号 (例: 10) を持つ行を抜き出す、ep番号は1から始まる
    specific_episode = stn_result[stn_result['episode'] == ep]

    # 1x6のデータフレームを作成
    stn_heatmap_data = pd.DataFrame(index=range(2, 3), columns=range(2, 8))

    # lc_hyperijのデータをheatmap_dataに挿入
    for j in range(2, 8):
        i = 2
        column_name = f'max_lc_hyper{i}{j}'
        stn_heatmap_data.at[i, j] = specific_episode[column_name].values[0]
        if stn_heatmap_data.at[i,j]>=1:
            stn_heatmap_data.at[i,j] -= 1

    # データを数値型に変換（必要な場合）
    stn_heatmap_data = stn_heatmap_data.astype(float)
    stn_heatmap_data = np.where(stn_heatmap_data==0, epsilon, stn_heatmap_data)
    plt.figure(figsize=(6, 5))
    stn_heatmap = plt.imshow(stn_heatmap_data, norm=SymLogNorm(linthresh=1e-10,vmin=0,vmax=0.2), cmap='Greens')

    # カラーバーの追加
    cbar = plt.colorbar(stn_heatmap)
    plt.yticks([0], ['0'])
    plt.savefig('stn_ep='+str(ep)+'.svg')
    plt.show()

for ep in range(2, 9, 2):
    """D1"""
    print(f'D1 ep={ep}')
    # 特定のepisode番号 (例: 10) を持つ行を抜き出す、ep番号は1から始まる
    specific_episode = d1_result[d1_result['episode'] == ep]

    # 1x6のデータフレームを作成
    d1_heatmap_data = pd.DataFrame(index=range(2, 3), columns=range(2, 8))

    # lc_hyperijのデータをheatmap_dataに挿入
    for j in range(2, 8):
        i = 2
        column_name = f'max_D1{i}{j}'
        d1_heatmap_data.at[i, j] = specific_episode[column_name].values[0]
        if d1_heatmap_data.at[i,j]>=1:
            d1_heatmap_data.at[i,j] -= 1

    # データを数値型に変換（必要な場合）
    d1_heatmap_data = d1_heatmap_data.astype(float)
    d1_heatmap_data = np.where(d1_heatmap_data==0, epsilon, d1_heatmap_data)

    plt.figure(figsize=(6, 5))
    d1_heatmap = plt.imshow(d1_heatmap_data, norm=SymLogNorm(linthresh=1e-10,vmin=0,vmax=0.2), cmap='Greens')

    # カラーバーの追加
    cbar = plt.colorbar(d1_heatmap)
    plt.yticks([0], ['0'])
    plt.savefig('d1_ep='+str(ep)+'.svg')
    plt.show()


for ep in range(2, 9, 2):
    """D2"""
    print(f'D2 ep={ep}')
    # 特定のepisode番号 (例: 10) を持つ行を抜き出す、ep番号は1から始まる
    specific_episode = d2_result[d2_result['episode'] == ep]

    # 1x6のデータフレームを作成
    d2_heatmap_data = pd.DataFrame(index=range(2, 3), columns=range(2, 8))

    # lc_hyperijのデータをheatmap_dataに挿入
    for j in range(2, 8):
        i = 2
        column_name = f'max_D2{i}{j}'
        d2_heatmap_data.at[i, j] = specific_episode[column_name].values[0]
        if d2_heatmap_data.at[i,j]<=1 and d2_heatmap_data.at[i,j]!=0:
            d2_heatmap_data.at[i,j] -= 1

    # データを数値型に変換（必要な場合）
    d2_heatmap_data = d2_heatmap_data.astype(float)
    d2_heatmap_data = np.where(d2_heatmap_data==0, epsilon, d2_heatmap_data)
    plt.figure(figsize=(6, 5))

    d2_heatmap = plt.imshow(d2_heatmap_data, norm=SymLogNorm(linthresh=1e-10,vmin=-0.2,vmax=0), cmap='Greens')

    # カラーバーの追加
    cbar = plt.colorbar(d2_heatmap)
    plt.yticks([0], ['0'])
    plt.savefig('d2_ep='+str(ep)+'.svg')
    plt.show()
```

### SNc activity

```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams["font.size"] = 18

da_log_all = []
ep_list = [10,50,100,150]
num_ep = len(ep_list)
num_unit = 6

# for i in range(num_ep):
for ep in ep_list:
    da_log_ep = []
    for j in range(2, 8):
        da_log_ep.append(file_name.groupby(['episode'])['DA{}{}'.format(2,j)].max()[ep])    
    da_log_all.append(da_log_ep)
    
prob_data = [np.array(row) / sum(row) for row in da_log_all]

# 5行のグラフを作成
fig, axes = plt.subplots(nrows=num_ep, ncols=1, figsize=(5, 10))

# 各行のデータをプロット
for i in range(num_ep):
    bars = axes[i].bar(range(num_unit), prob_data[i], width=1)  # 各行のデータをバーグラフとしてプロット
    axes[i].set_ylim(0, 0.35)                   # y軸の範囲を0~0.35に設定
    axes[i].set_xticks(range(num_unit))                # x軸の目盛りを設定
    axes[i].set_xticklabels([f'{j}' for j in range(6)])  # x軸ラベルを設定
    axes[i].set_xlim(-0.5,5.5)
    axes[i].set_ylim(0,1.0)
    axes[i].set_title('ep = {}'.format(ep_list[i]))    

    # 軸の太さを変更
    for spine in axes[i].spines.values():
        spine.set_linewidth(1)  # ここで太さを指定（例: 2）

# レイアウトを整える
plt.tight_layout()
plt.savefig('Comparison_with_schultz_experiment.svg')
plt.show()
```
